```python
# -----------------------------------------------------------------------------
# SCRIPT 2: PROFESSIONAL DASHBOARD GENERATOR (DRIVE-INTEGRATED)
# -----------------------------------------------------------------------------
import pandas as pd
import plotly.express as px
from google.colab import drive, files

# STEP 1: Mount Google Drive to access the cleaned sample
drive.mount('/content/drive', force_remount=True)
FILE_PATH = '/content/drive/MyDrive/Colab Notebooks/cleaned_sampled_data.csv'

# STEP 2: Load the high-quality cleaned dataset
print(f"🚀 Loading cleaned data from: {FILE_PATH}")
df = pd.read_csv(FILE_PATH)

# Professional Color Palette
COLOR_THEME = px.colors.qualitative.Pastel

def create_final_presentation_dashboard():
    # --- CHART 1: Burnout Density (Violin Plot) ---
    # Visualizes the probability density distribution across demographics
    fig1 = px.violin(df, y="burnout_score", x="gender", color="gender", box=True,
                     title="1. Burnout Density Distribution (Statistical Density)",
                     color_discrete_sequence=COLOR_THEME, template="plotly_white")

    # --- CHART 2: Stressor Analysis (Grouped Bar) ---
    # Aggregates stressors and sorts from High to Low risk for logical flow
    stress_metrics = ['exam_pressure', 'financial_stress', 'family_expectation']
    grouped_data = df.groupby('risk_level', observed=True)[stress_metrics].mean().reset_index()

    fig2 = px.bar(grouped_data, x="risk_level", y=stress_metrics, barmode="group",
                  title="2. Stressor Analysis by Risk Level (High to Low Trend)",
                  labels={'value': 'Avg Score', 'variable': 'Stressor Type'},
                  color_discrete_sequence=COLOR_THEME, template="plotly_white")

    # Explicit sorting logic to ensure High -> Medium -> Low progression
    fig2.update_xaxes(categoryorder='array', categoryarray=['High', 'Medium', 'Low'])

    # --- CHART 3: Global Correlation Matrix (Heatmap) ---
    # Justifies feature selection for the upcoming Machine Learning phase
    corr_matrix = df.select_dtypes('number').corr()
    fig3 = px.imshow(corr_matrix, text_auto=".2f",
                    title="3. Global Correlation Matrix (Feature Interaction Logic)",
                    color_continuous_scale='RdBu_r', aspect="auto", template="plotly_white")

    # STEP 3: Visual Refinements & Legend Positioning
    # Moving legends to the bottom to maximize charting area and prevent title overlap
    for f in [fig1, fig2]:
        f.update_layout(
            font=dict(size=12),
            legend=dict(orientation="h", y=-0.25, x=0.5, xanchor="center"),
            margin=dict(t=80, l=50, r=50, b=120)
        )
    # Heatmap specific margin for label clarity on the Y-axis
    fig3.update_layout(margin=dict(t=80, l=120, r=50, b=120))

    # STEP 4: Generate Professional HTML Grid Layout
    html_layout = f"""
    <html>
    <head>
        <title>Student Burnout Analysis - Graduate Study</title>
        <style>
            body {{ font-family: 'Segoe UI', sans-serif; background-color: #f8f9fa; margin: 0; padding: 30px; }}
            .header {{ text-align: center; padding: 25px; background: linear-gradient(135deg, #1e3c72, #2a5298); color: white; border-radius: 15px; margin-bottom: 30px; box-shadow: 0 5px 15px rgba(0,0,0,0.2); }}
            .grid {{ display: grid; grid-template-columns: repeat(2, 1fr); gap: 30px; max-width: 1600px; margin: auto; }}
            .card {{ background: white; border-radius: 15px; padding: 20px; box-shadow: 0 10px 20px rgba(0,0,0,0.08); height: 550px; overflow: hidden; }}
            /* Layout: Charts 1 & 2 side-by-side, Chart 3 spanning full width at bottom */
            .full-width {{ grid-column: span 2; height: 600px; }}
            h1 {{ margin: 0; font-size: 32px; letter-spacing: 1px; }}
            p {{ margin-top: 5px; opacity: 0.85; font-weight: 300; }}
        </style>
    </head>
    <body>
        <div class="header">
            <h1>🎓 Student Well-being Analytics Dashboard</h1>
            <p>Predictive EDA on 1,000,000 Records | Processing via NVIDIA RAPIDS</p>
        </div>
        <div class="grid">
            <div class="card">{fig1.to_html(full_html=False, include_plotlyjs='cdn')}</div>
            <div class="card">{fig2.to_html(full_html=False, include_plotlyjs=False)}</div>
            <div class="card full-width">{fig3.to_html(full_html=False, include_plotlyjs=False)}</div>
        </div>
    </body>
    </html>
    """

    # Export the final interactive HTML file
    FILENAME = "Final_Burnout_Analysis_3Charts.html"
    with open(FILENAME, "w", encoding="utf-8") as f:
        f.write(html_layout)
    print(f"✨ Dashboard Successfully Generated: '{FILENAME}'")

# Execute the generator
create_final_presentation_dashboard()

# Trigger browser download
files.download("Final_Burnout_Analysis_3Charts.html")
```
