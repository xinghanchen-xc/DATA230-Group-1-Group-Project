```python
# -----------------------------------------------------------------------------
# SCRIPT 1: DATA CLEANING & STATISTICAL SAMPLING (GPU POWERED)
# -----------------------------------------------------------------------------
import cudf
import time
from google.colab import drive, files

# STEP 1: Mount Google Drive to access the large dataset
drive.mount('/content/drive', force_remount=True)
FILE_PATH = '/content/drive/MyDrive/Colab Notebooks/student_mental_health_burnout_1M.csv'

# STEP 2: Load 1,000,000 rows directly into GPU memory using cuDF
print("🚀 Initializing GPU-accelerated data loading...")
start_time = time.time()
gdf = cudf.read_csv(FILE_PATH)

# STEP 3: Basic Preprocessing
# Standardize column names (lowercase and strip whitespace)
gdf.columns = gdf.columns.str.strip().str.lower()

# Remove duplicate records and drop any missing values (Ensuring Data Integrity)
gdf = gdf.drop_duplicates().dropna()

# STEP 4: Advanced Statistical Cleaning (Z-Score Outlier Removal)
# We remove outliers (> 3 Standard Deviations) to ensure the EDA trends are not skewed
for col in ['academic_performance', 'burnout_score', 'sleep_hours']:
    mean_val = gdf[col].mean()
    std_val = gdf[col].std()
    # Apply boolean indexing filter on GPU
    gdf = gdf[(gdf[col] >= mean_val - 3 * std_val) & (gdf[col] <= mean_val + 3 * std_val)]

# STEP 5: Sampling for Visualization Performance
# We extract 50,000 records. This sample is statistically significant as it
# is derived from a cleaned 1M-row population, ensuring smooth browser rendering.
df_final = gdf.sample(n=50000).to_pandas()

# STEP 6: Export Cleaned Sample
OUTPUT_NAME = "cleaned_sampled_data.csv"
df_final.to_csv(OUTPUT_NAME, index=False)

end_time = time.time()
print(f"✅ Data processing complete in {end_time - start_time:.4f} seconds.")
print(f"📉 Final Dataset Shape: {df_final.shape} (Processed from 1,000,000 rows)")

# Trigger browser download for the cleaned file
files.download(OUTPUT_NAME)
```
